```python
import os
import numpy as np
import math
import shutil

def split_data(data_dir, train_ratio, val_ratio):
    # Define the root directory of your dataset
    ROOT_DIR = data_dir

    # Create directories for training, validation, and testing
    train_dir = os.path.join(ROOT_DIR, "train")
    val_dir = os.path.join(ROOT_DIR, "val")
    test_dir = os.path.join(ROOT_DIR, "test")

    os.makedirs(train_dir, exist_ok=True)
    os.makedirs(val_dir, exist_ok=True)
    os.makedirs(test_dir, exist_ok=True)

    # Iterate through subdirectories (classes)
    for class_dir in os.listdir(ROOT_DIR):
        class_path = os.path.join(ROOT_DIR, class_dir)

        # Check if the path is a directory
        if os.path.isdir(class_path):
            # Count the number of images in the class directory
            image_count = len(os.listdir(class_path))

            # Calculate the number of images for each split
            num_train_images = math.floor(train_ratio * image_count)
            num_val_images = math.floor(val_ratio * image_count)
            num_test_images = image_count - num_train_images - num_val_images

            # Create subdirectories for this class in train, val, and test
            train_class_dir = os.path.join(train_dir, class_dir)
            val_class_dir = os.path.join(val_dir, class_dir)
            test_class_dir = os.path.join(test_dir, class_dir)

            os.makedirs(train_class_dir, exist_ok=True)
            os.makedirs(val_class_dir, exist_ok=True)
            os.makedirs(test_class_dir, exist_ok=True)

            # List all images in the class directory
            class_images = [img for img in os.listdir(class_path) if os.path.isfile(os.path.join(class_path, img))]
            # Shuffle the list to randomize
            np.random.shuffle(class_images)

            # Copy images to the respective directories
            for i, img in enumerate(class_images):
                source_path = os.path.join(class_path, img)
                if i < num_train_images:
                    destination_path = os.path.join(train_class_dir, img)
                elif i < num_train_images + num_val_images:
                    destination_path = os.path.join(val_class_dir, img)
                else:
                    destination_path = os.path.join(test_class_dir, img)
                shutil.copy(source_path, destination_path)

    print("Data split into train, validation, and test sets")

# Example usage:
split_data("/content/tea sickness dataset", 0.7, 0.15)
```

```python
import os
import numpy as np
import math
import shutil

def dataSplit(data_dir, train_split, val_split, test_split):
    # Define the root directory of your dataset
    ROOT_DIR = data_dir

    # Create subdirectories for train, validation, and test
    for subset in ["train", "val", "test"]:
        os.makedirs(subset, exist_ok=True)

    # Iterate through subdirectories (classes)
    for class_dir in os.listdir(ROOT_DIR):
        class_path = os.path.join(ROOT_DIR, class_dir)

        # Check if the path is a directory
        if os.path.isdir(class_path):
            class_images = [img for img in os.listdir(class_path) if img.lower().endswith(('.jpg', '.jpeg', '.png', '.bmp'))]
            num_images = len(class_images)

            # Calculate the number of images to copy for each subset
            num_train_images = math.floor(train_split * num_images)
            num_val_images = math.floor(val_split * num_images)
            num_test_images = num_images - num_train_images - num_val_images

            # Randomly shuffle the images
            np.random.shuffle(class_images)

            # Define destination paths for each subset
            train_class_dir = os.path.join("train", class_dir)
            val_class_dir = os.path.join("val", class_dir)
            test_class_dir = os.path.join("test", class_dir)

            # Ensure destination directories are created
            os.makedirs(train_class_dir, exist_ok=True)
            os.makedirs(val_class_dir, exist_ok=True)
            os.makedirs(test_class_dir, exist_ok=True)

            # Copy images to the corresponding subsets
            for img in class_images[:num_train_images]:
                source_path = os.path.join(class_path, img)
                destination_path = os.path.join(train_class_dir, img)
                shutil.copy(source_path, destination_path)

            for img in class_images[num_train_images:num_train_images + num_val_images]:
                source_path = os.path.join(class_path, img)
                destination_path = os.path.join(val_class_dir, img)
                shutil.copy(source_path, destination_path)

            for img in class_images[-num_test_images:]:
                source_path = os.path.join(class_path, img)
                destination_path = os.path.join(test_class_dir, img)
                shutil.copy(source_path, destination_path)

    print("Data split into train, validation, and test sets")

# Example usage:
dataSplit("/content/tea sickness dataset", 0.7, 0.15, 0.15)
```

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten

# Data preprocessing and augmentation
train_data_gen = tf.keras.preprocessing.image.ImageDataGenerator(
    rescale=1.0 / 255.0,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True
)

val_test_data_gen = tf.keras.preprocessing.image.ImageDataGenerator(rescale=1.0 / 255.0)

train_data = train_data_gen.flow_from_directory(
    train_dir,
    target_size=(224, 224),
    batch_size=32,
    class_mode='categorical'  # Change this to 'categorical' if you have multiple classes
)

val_data = val_test_data_gen.flow_from_directory(
    val_dir,
    target_size=(224, 224),
    batch_size=32,
    class_mode='categorical'  # Change this to 'categorical' if you have multiple classes
)

# Model architecture
model = Sequential()

# Add a Flatten layer to convert the data for feeding into a fully connected neural network
model.add(Flatten(input_shape=(224, 224, 3)))

# Add Dense layers for your neural network architecture
model.add(Dense(128, activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dense(32, activation='relu'))

# Output layer
model.add(Dense(1, activation='sigmoid'))  # Output layer for binary classification

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Early stopping and model checkpoint
es = tf.keras.callbacks.EarlyStopping(monitor="val_accuracy", min_delta=0.01, patience=3, verbose=1, mode='auto')
mc = tf.keras.callbacks.ModelCheckpoint("bestmodel.h5", monitor="val_accuracy", verbose=1, save_best_only=True, mode='auto')

cd = [es, mc]  # Define the list of callbacks

# Model training
hs = model.fit(
    train_data,
    steps_per_epoch=len(train_data),
    epochs=30,
    verbose=1,
    validation_data=val_data,
    validation_steps=len(val_data),
    callbacks=cd  # Pass the callbacks to the fit method
)
```

```python
import matplotlib.pyplot as plt

# Assuming 'h' contains your model's training history
h = hs.history

# Extract the keys from the history dictionary
keys = h.keys()
print(keys)

# Plot training loss
plt.plot(h['loss'], label='Training Loss')
plt.plot(h['val_loss'], label='Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.title('Training and Validation Loss')
plt.show()

# Plot training accuracy
plt.plot(h['accuracy'], label='Training Accuracy')
plt.plot(h['val_accuracy'], label='Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.title('Training and Validation Accuracy')
plt.show()
```

```python
import numpy as np
from keras.models import Sequential
from keras.layers import Flatten, Dense
from keras.preprocessing.image import ImageDataGenerator

# Define your data directories
train_dir = "/content/train"
val_dir = "/content/val"
test_dir = "/content/test"

# Data preprocessing and augmentation
train_data_gen = ImageDataGenerator(
    rescale=1.0 / 255.0,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True
)

val_test_data_gen = ImageDataGenerator(rescale=1.0 / 255.0)

train_data = train_data_gen.flow_from_directory(
    train_dir,
    target_size=(224, 224),
    batch_size=32,
    class_mode='categorical'  # Change this to 'categorical' if you have multiple classes
)

val_data = val_test_data_gen.flow_from_directory(
    val_dir,
    target_size=(224, 224),
    batch_size=32,
    class_mode='categorical'  # Change this to 'categorical' if you have multiple classes
)

# Model architecture
model = Sequential()

model.add(Flatten(input_shape=(224, 224, 3)))  # Input layer for ANN

model.add(Dense(128, activation='relu'))  # Hidden layer
model.add(Dense(64, activation='relu'))   # Hidden layer

model.add(Dense(1, activation='sigmoid'))  # Output layer for binary classification

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Model training
hs = model.fit(
    train_data,
    steps_per_epoch=len(train_data),
    epochs=30,
    verbose=1,
    validation_data=val_data,
    validation_steps=len(val_data)
)

# Model Accuracy
test_data_gen = ImageDataGenerator(rescale=1.0 / 255.0)

test_data = test_data_gen.flow_from_directory(
    test_dir,
    target_size=(224, 224),
    batch_size=32,
    class_mode='categorical'  # Change this to 'categorical' if you have multiple classes
)

acc = model.evaluate(test_data)[1]

print(f"The accuracy of our model is {acc * 100:.2f}%")
```
